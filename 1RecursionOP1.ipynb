## Principle of Mathematical Induction

Let us considering P(n) = 1 + 2 + 3 + ….. + n = n(n +1)/2 
For the n = 1 
LHS of P (n) = 1 
RHS of P (n) = 1 
(1+1)/2 = 1 
Therefore, LHS = RHS 
Here, P (n) is true for n = 1 
Let us consider P (n) be the true for n = k, 
therefore 1 + 2 + 3 + …. + k = k (k+1)/2 … (i) 
Then, 
(1 + 2 + 3 + … + k) + (k + 1) = k(k + 1)/2 + (k + 1) 
= (k + 1) (k/2 + 1) 
= [(k + 1) (k + 2)]/2 
= [(k + 1) [(k + 1) + 1]]/2 
P (n) is true for n = k + 1 
P (n) is true for all n ∈ N 
Therefore, by the principle of Mathematical Induction 
Thus, P (n) = 1 + 2 + 3 + ….. + n = n (n +1)/2 is true for all n ∈ N.

### Factorial using Recursion

```python
def factorial(n):
    if n==0 or n==1:               #Trivial case/Base case
        return 1
    smallOutput = sum(n-1)        #Induction step
    return n * smallOutput
```

### Sum of natural numbers using recursion

```python
def sum(n):
    if n==1:
        return 1
    smallOutput = sum(n-1)
    return n + smallOutput
```

```python
n = int(input())
print(factorial(n))
print(sum(n))
```

### Fibonacci Series

```python
def fib(n):
    if n==1 or n==2:
        return 1
    fib_n_1 = fib(n-1)
    fib_n_2 = fib(n-2)
    output = fib_n_1 + fib_n_2
    return output
n=int(input())
print(fib(n))
```

### To check wheter the given list is sorted or not

```python
def isSorted(a):
    l = len(a)
    if l==0 or l==1:
        return True
    if a[0]>a[1]:
        return False
    smallerlist = a[1:]
    isSmallerListSorted = isSorted(smallerlist)
    return isSmallerListSorted

    
```

```python
a = [1,2,3,4,5,6,7,8,9]
isSorted(a)
```

```python
a = [1,2,3,14,5,6,7,8,9]
isSorted(a)
```

The problem with above function is that, it requires a lot of space (smallerlist = a[1:], suppose a list with 1000 elements, firstly it will create a copy of 999 elements,998 and so on...)

```python
#improved function
def isSortedBetter(a,si):            #si=startIndex
    l = len(a)
    if si==l-1 or si==l:
        return True
    if a[si]>a[si+1]:
        return False
    isSmallerPartSorted = isSortedBetter(a,si+1)
    return isSmallerPartSorted
a = [1,2,3,14,5,6,7,9]
si = 0
isSortedBetter(a,si)

    
```

### Sum of Array

Taking first element and adding it to remaining list

```python
def sumarray(a):
    l = len(a)
    if l==0:
        return 0
    if l==1:
        return a[0]
    smallerList = a[1:]
    sum = a[0] + sumarray(smallerList)
    return sum
a = [1,2,3,4,5]
sumarray(a)
```

Taking last element and adding it to remaining list

```python
def sumarray(a):
    l = len(a)
    if l==0:
        return 0
    if l==1:
        return a[0]
    smallerList = a[0:l-1]
    sum = a[l-1] + sumarray(smallerList)
    return sum
a = [1,2,3,4,5]
sumarray(a)
```

Breaking the list in between

```python
def sumarray(a):
    l = len(a)
    if l==0:
        return 0
    if l==1:
        return a[0]
    smallerList = a[0:(l//2)]
    sum = sumarray(a[l//2:]) + sumarray(smallerList)
    return sum
a = [1,2,3,4,5]
sumarray(a)
```

### Checking a number in an array

```python
def check(a,n):
    l = len(a)
    if l==0:
        return False
    if n==a[0]:
        return True
   
    smallerList = a[1:]
    p = check(smallerList,n)
   
    return p
a = [1,2,3,4,5,6,7]
n = 1
check(a,n)
```

```python
def check(a,n):
    l = len(a)
    if l==0:
        return False
    if n==a[l-1]:
        return True
   
    smallerList = a[0:l-1]
    p = check(smallerList,n)
   
    return p
a = [1,2,3,4,5,6,7]
n = 10
check(a,n)
```

### First Occurence of a number in a List

Using Start Index

```python
def firstIndex(a,x,si):
    l = len(a)
    if si==l:
        return -1
    
    if a[si] == x:
        return si
    p = firstIndex(a,x,si+1)
    return p
a = [1,1,2,4,1,5,1]
firstIndex(a,7,0)
    
        
```

Using Copy

```python
def firstIndex(a,x):
    l = len(a)
    if l==0:
        return -1
    if a[0] == x:
        return 0
    smallerList = a[1:]
    p = firstIndex(smallerList,x)
    if p==-1:
        return -1
    else:
        return p+1
    
a = [1,1,2,4,1,5,1]
print(firstIndex(a,2))
```

### Last Occurence of a number in a List

Using Copy

```python
def lastIndex(a,x):
    l = len(a)
    if l == 0:
        return -1
    smallerList=a[1:]
    p = lastIndex(smallerList,x)
    
    if p==-1:
        if a[0]==x:
            return 0
        return -1
    else:
        return p+1
a = [1,1,2,4,1,5,1]
print(lastIndex(a,1))
```

Using Start Index

```python
def lastIndex(a,x,si):
    l = len(a)
    if si ==l:
        return -1
    p = lastIndex(a,x,si+1)
    
    if p==-1:
        if a[si]==x:
            return 0
        return -1
    else:
        return p+1
a = [1,1,2,4,1,5,1]
print(lastIndex(a,2,0))
```
